```python
import pandas as pd
import numpy as np
import keras
```

```python
# UCI HAR Dataset
x_train = pd.read_table("UCI HAR Dataset\\train\\X_train.txt",delim_whitespace=True,header=None)
```

```python
x_train
```

```python
y_train = pd.read_table("UCI HAR Dataset\\train\\Y_train.txt",header=None)
```

```python
y_train
```

```python
x_test = pd.read_table("UCI HAR Dataset\\test\\X_test.txt",delim_whitespace=True,header=None)
y_test = pd.read_table("UCI HAR Dataset\\test\\Y_test.txt",header=None)
```

```python
type(x_train)
```

```python
# convert dataframe into array
x_train = x_train.values
y_train = y_train.values
print(x_train.shape)
print(y_train.shape)
```

```python
# One-Hot Encoding
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
y_train = ohe.fit_transform(y_train).toarray()
```

```python
print(y_train)
print(y_train.shape)
```

```python
# Artificial Neural Network
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout  
```

```python
model = Sequential()
model.add(Dense(48, input_dim = 561, kernel_initializer='uniform', activation='relu', ))
model.add(Dropout(0.1))
model.add(Dense(24, kernel_initializer='uniform', activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(12, kernel_initializer='uniform', activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(6, kernel_initializer='uniform', activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics = ['accuracy'])
```

```python
# Train
model.fit(x_train, y_train, batch_size=20, epochs=10, verbose = 4)
```

```python
# Training and Testing Accuracy
train_acc = model.evaluate(x_train, y_train)
print("Training Accuracy:", train_acc[1])
y_test = ohe.fit_transform(y_test).toarray()
test_acc = model.evaluate(x_test, y_test)
print("Testing Accuracy:", test_acc[1])
```
